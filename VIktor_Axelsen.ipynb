<b>SERVICE ERRORS</b>

These are the four main types of services in badminton and most can be executed with either your forehand or backhand.
<br>Low serve. 
<br>High serve. 
<br>Flick serve. 
<br>Drive Serve.

Service is a start of any Game or a play. Taking into consideration his international ranking, the service errors must be minimal against any opponent.

```python
#SERVICE_ERROR with respect to opponent
C=df.loc[:,['OPPONENT','RESULT','SERVICE_ERROR']]
C.sort_values(by=['SERVICE_ERROR'], ascending=False)
```

1.From the above analysis we see that were Service Errors are less than or equal to 2, Mr Axelsen has Won more games.<br><br>2.But When Service Errors are more than 3, Mr. Axelsen has Lost more number of games.<br><br>

This Representation of Analysis is shown in graph.

```python
plt.figure(figsize=(8, 10))
sns.scatterplot(data=C, x="SERVICE_ERROR", y="OPPONENT").set(title='Service Error with respect to opponent')
plt.show()
```

From above Visualization, we conclude that the Service Error range is between 2 and 3 against any Opponent.

<b>------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------</b>

<b>STRAIGHT TOSS ERROR</b>

```python
#STRAIGHT_TOSS_ERROR
fig, ax = plt.subplots(figsize=(10,10))
sns.swarmplot(x='STRAIGHT_TOSS_ERROR', y='OPPONENT', data=df,size=10).set(title='STRAIGHT TOSS ERROR')
plt.show()
```

<b>The above graphs indicates the analysis on Straight Toss Error</b>

As you can see for Straight Toss Error, Maximum is in the range of 2 and 3 against any opponent played.

<b>CROSS TOSS ERROR</b>

```python
#CROSS_TOSS__ERROR
fig, ax = plt.subplots(figsize=(10,10))
sns.swarmplot(x='CROSS_TOSS__ERROR', y='OPPONENT', data=df,size=10).set(title='CROSS TOSS ERROR')
plt.show()
```

<b>The above graphs indicates the analysis on Cross Toss Error</b>

Similar to Straight Toss Error, Cross Toss Error has Maximum range between 2 and 3 against any opponent played.

<b>CONCLUSION:</b><br>Comparing the Straight Toss Error and Cross Toss Error Mr. Axelsen need to work more on <b>Cross Toss Error</b> as the maximum error range is between 3 and 4.

Similar Analysis can be done for other Errors.

```python
# Smash Return Error against Opponents
plt.figure(figsize=(8, 10))
ax = sns.barplot(x="SMASH_RETURN_ERROR", y="OPPONENT", data=H,ci=None,hue='RESULT').set(title='Smash Return Error Against Opponents')
```
